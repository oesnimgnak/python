```python
#라이브러리 불러오기
from wordcloud import WordCloud # 워드클라우드 생성을 위한 라이브러리
import matplotlib.pyplot as plt # 그래프 및 시각화를 위한 라이브러리
from collections import Counter # 리스트 요소의 개수를 셀 때 사용하는 라이브러리
from konlpy.tag import Okt  # 한국어 테스트를 형태소로 분석하기 위한 라이브러리
import numpy as np  # 배열 및 행렬 연산을 위한 라이브러

okt = Okt() # 형태소 분석기 초기화
```

```python
# 텍스트 분석하기

# 파일 오픈
with open('대한민국헌법.txt','r',encoding='UTF-8') as f :
    text = f.read()

# 형태소 분서을 통해 명사 추출
nouns = okt.nouns(text)

# 한 글자 이상의 명사만 출출
words = [n for n in nouns if len(n) > 1]

# 단어의 빈도수 계산
c = Counter(words)

# 코랩에서 제공하는 한글 폰트 사용을 위해 설정합니다.
#plt.rc('font', family = 'NanumBarunGothic')
#font_path = '/usr/share/fonts/truetype/nanum/NanumBarunGothic.ttf'

# 위드클라우드 생성을 위한 설정.
font_path = '/content/H2MKPB.TTF'   # 업로드한 폰트
wc = WordCloud(width=400, height=400, font_path=font_path, max_font_size=250)

# 단어의 빈도수에 따라 워드클라우드 생성
gen = wc.generate_from_frequencies(c)

# 워드클라우드 시각화 및 이미지 파일로 저장
plt.imshow(gen)
# 그래프를 화면에 출력
plt.show()
wc.to_file('헌법_클라우드.png')
```
